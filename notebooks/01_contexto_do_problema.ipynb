# Análise de Gastos Mensais

## 01 - Contexto do Problema
O controle e a análise dos gastos mensais são fundamentais para compreender o comportamento financeiro individual e apoiar a tomada de decisões mais conscientes. No dia a dia, as despesas se distribuem entre diferentes categorias e variam ao longo dos meses, o que dificulta a identificação de padrões apenas por observação direta. Este projeto analisa gastos mensais a partir de um conjunto de dados fictícios, utilizando técnicas de estatística descritiva, com o objetivo de compreender o comportamento de consumo e identificar possíveis oportunidades de otimização financeira.

## Cenário do Projeto
* Os dados utilizados neste projeto são fictícios e representam os gastos mensais de uma pessoa física ao longo de 12 meses (Janeiro a Dezembro).
* A construção do dataset considerou uma renda mensal próxima à média nacional, estimada em 3.208 reais em 2024, segundo o IBGE.
* Com isso, os valores de gastos foram ajustados para totalizar, em média, cerca de 3.000 reais por mês, refletindo uma realidade em que grande parte da renda é comprometida com despesas essenciais.

## Objetivos da Análise

Os principais objetivos deste projeto são:

- Analisar os gastos mensais a partir de um conjunto de dados fictícios.
- Aplicar conceitos de estatística descritiva na análise dos dados.
- Identificar medidas de tendência central (média, mediana e moda).
- Avaliar a dispersão dos dados por meio de variância e desvio padrão.
- Analisar a distribuição dos gastos entre diferentes categorias.
- Investigar variações dos gastos ao longo dos meses.
- Extrair insights que auxiliem no controle e planejamento financeiro.

## Organização do Projeto

A análise foi estruturada em múltiplos notebooks, cada um responsável por uma etapa do processo:

- Contexto do problema
- Entendimento dos dados
- Análise exploratória
- Estatística descritiva
- Interpretação dos resultados
- Conclusão

Nos próximos notebooks, os dados serão explorados e analisados utilizando tabelas, visualizações gráficas e medidas estatísticas.
